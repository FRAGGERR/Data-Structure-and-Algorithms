### Returning the Index

```python
def returnIndex(arr, search):
    for index in range(len(arr)):
        if arr[index] == search:
            return index
    return -1

if __name__ == '__main__':
    
    arr = [3,5,23,66,34,56,25,79]
    search = 25
    index = returnIndex(arr, search)
    print(index)
```

### Search In Range i.e., [1, 5]

```python
#Inputs
arr = [2, 36, 12, 97, -13, 61, -46]
target = -13

def searchInRange(arr, target, start, end):
    if len(arr) == 0: #if the given list is empty
        return 'Empty List'
        
    for i in range(start, end+1): #checking in the range
        if arr[i] == target:
            return i

    return -1

#output
if __name__ == '__main__':
    print(searchInRange(arr, target, 1, 5))
```

```python
arr = [10, 23, 35, 45, 50 ,70, 85]
target = 70

def binarysearch(arr, target):
    start = 0
    end = len(arr) - 1
    
    while(start <= end):
        mid = (start + end)//2
        
        if (arr[mid] < target):
            start = mid + 1
        elif (arr[mid] > target):
            end = mid - 1
        else:
            return mid, arr[mid] #give the index and the real value

    return -1

if __name__ == '__main__':
    print(binarysearch(arr, target))
```

•Best-case:  O(1)  — The target is found at the midpoint in the first comparison.

•Average-case:  O(log n)  — On average, it takes  \log n  comparisons to find the target.

•Worst-case:  O(log n)  — The worst case occurs when the search continues through all halves before determining the target’s location or its absence.

## Recursive way Binary Search

```python
arr = [10, 23, 35, 45, 50 ,70, 85]
target = 85

def recursiveBinarySearch(arr, target, start, end):
    if start > end:
        return -1
    mid = (start + end)//2
    
    if arr[mid] == target:
        return mid, arr[mid]
    elif arr[mid] > target:
        return recursiveBinarySearch(arr, target, start, mid - 1)
    else:
        return recursiveBinarySearch(arr, target, mid + 1, end)

if __name__ == '__main__':
    print(recursiveBinarySearch(arr, target, 0, len(arr) -1))
```

    •Best-case:  O(1)  — The target is found at the midpoint on       the first comparison.
    
	•Average-case:  O(log n)  — On average, the algorithm will        halve the search space  log n  times.
    
	•Worst-case:  O(log n)  — The worst case occurs when the          algorithm has to search through all halved sections of the       array before finding the target (or determining it’s not         present).
